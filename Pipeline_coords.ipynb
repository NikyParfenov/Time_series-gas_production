```python
result = pd.DataFrame(columns=['Pipeline', 'Latitude', 'Longitude'])
for i in range(len(gj['features'])):
    if len(gj['features'][i]['geometry']['coordinates']) > 2:
        _result = pd.DataFrame(data=gj['features'][i]['geometry']['coordinates'],
                               columns=['Latitude', 'Longitude']
                              )
        if gj['features'][i]['properties']['clusterCaption'] not in not_used:
            _result['Pipeline'] = gj['features'][i]['properties']['clusterCaption']
            result = result.append(_result)

profiles = len(result['Pipeline'].unique())
print(f'Всего {profiles} газопровода(ов)')
```
